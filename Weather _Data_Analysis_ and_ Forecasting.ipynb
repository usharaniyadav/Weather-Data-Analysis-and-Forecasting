```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\WINDOWS\OneDrive\Documents\nasheit1\Student Performance Data Analysis\weather_dataset.csv")
```

```python

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['Temperature'].plot()
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
df['Rainfall'].resample('ME').sum().plot(kind='bar')
plt.title("Monthly Rainfall")
plt.show()
```
