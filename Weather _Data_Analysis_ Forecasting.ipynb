**BAR GRAPH (Monthly Rainfall)**

```python

df = pd.read_csv("weather_dataset.csv")
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['Rainfall'].resample('ME').sum().plot(kind='bar')
plt.title("Monthly Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.show()
df.reset_index(inplace=True)
```

**PIE CHART (Weather Condition)**

```python

df['Condition'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Weather Condition Distribution")
plt.ylabel("")
plt.show()
```

**HISTOGRAM (Temperature)**

```python
plt.hist(df['Temperature'])
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

**BOXPLOT (Temperature)**

```python
plt.boxplot(df['Temperature'])
plt.title("Temperature Boxplot")
plt.show()
```

 **SCATTER PLOT (Humidity vs Rainfall)**

```python
plt.scatter(df['Humidity'], df['Rainfall'])
plt.title("Humidity vs Rainfall")
plt.xlabel("Humidity")
plt.ylabel("Rainfall")
plt.show()
```
